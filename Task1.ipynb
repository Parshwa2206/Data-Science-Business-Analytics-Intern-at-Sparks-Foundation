```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
df.head(15)
```

```python
df.plot(x='Hours',y='Scores',style='*',color='green')
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['Hours']].values
Y = df[['Scores']].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size=0.7)
```

```python
X_train.shape , X_test.shape , Y_train.shape , Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
LinReg.fit(X_train, Y_train)

print("Your Model is trained now:)")
```

```python
plt.scatter(LinReg.predict(X_train), LinReg.predict(Y_train), 
            color = "green",label = 'Train data') ;
plt.xlabel('Hours');
plt.ylabel('Scores');
```

```python
line = LinReg.coef_*X + LinReg.intercept_

plt.scatter(X,Y)
plt.plot(X,line,color='red');
plt.show()
```

```python
Enter = float(input("Enter number of Hours you study : "))
print("Your Predicted Score : " , LinReg.predict([[Enter]]) [0][0] )
```

```python
print("Actual Value:\n",Y_test);
Y_pred = LinReg.predict(X_test);
print("Predicted value:\n",Y_pred);
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error;
print("Mean Squared Error:" , mean_squared_error(Y_test, Y_pred) );
print("Mean Absolute Error:" , mean_absolute_error(Y_test, Y_pred) );
```
